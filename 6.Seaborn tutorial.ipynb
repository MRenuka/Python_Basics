# Seaborn tutorial (Tutorial-9)

## Distribution plots
- distplot
- joinplot
- pairplot

```python
import seaborn as s
```

```python
df=s.load_dataset("tips")
```

```python
df.head() 
#consider scenario
#this is dataset from resturant and customers gave tip based on bill generation 
#so tip is our dependent feature
#other fields are independent features
```

# Correlation with heatmap

```python
#Heatmap is used to know corelation(inter relationship) between fields
#we can find correlation for numerical(int/float) values but not catogerical values
#corelation values ranges between -1 and +1
```

```python
df.corr() 
#as we said that correaltion is applied for int/float we got total_bill,tip,size fields only
```

```python
#if we want to see above values in graphical format then we use heatmap 
s.heatmap(df.corr())
```

# Joinplot
- allows us to study the relationship between 2 numeric values(features) called as uni-variate/bi-variate analysis

## Univariate analysis

```python
s.jointplot(x='tip',y='total_bill',data=df,kind='hex')
#kind='scatter'/ 'reg(regression)'/ 'resid'/ 'kde'/'hex(hexagon)'
```

# Pairplot

```python
#if features are more than 2 then we use pairplot it is also called as scatterplot
# fetaures can be int/float
# it will compute permutation and combination of feature values
s.pairplot(df)
```

```python
#considering 1 field and do pairplot this is possible through "hue"
s.pairplot(df,hue='sex')
#as sex consist of male,female there are 2 colors 
```

```python
# if we want value count of males and females
df['sex'].value_counts()
```

## Distplot
- helps to create hsitogram(based on x-axis,density is shown on y-axis)
- we can write only for 1 feature

```python
s.distplot(df['tip'])
```

```python
# in the above histogram on Y-axis density is show instead if we want values we need to give 
# kde(kernel density estimation)=false by default it is true
s.distplot(df['tip'],kde=False,bins=10)
```

```python
#Seaborn also helps in analyising on catogerical data
```

## CATOGERICAL PLOTS
- bar plot
- count plot
- violin plot
- box plot

```python
# count plot
# name itself says that we can see the count of particualr field(sex,smoker...etc)
# for count plot we can give only 1 field
# if we give more than 1 field then it will give error

s.countplot(x='sex',data=df)
```

```python
# if we want to change the view of graph towards 

s.countplot(y='sex',data=df)
```

```python
#In order to overcome the problem of countplot(as we can use only 1 feature) we can use 'Barplot'
```

## BarPlot

```python
#In barplot either x/y should have numeric field then only we can use barplot else it will give error

s.barplot(x='sex',y='tip',data=df)
```

```python
#Apart from this we can use hue(means based one hue parameter remaning parameters are classified)

s.barplot(x='sex',y='tip',data=df,hue='smoker')
```

## BoxPlot

```python
# box plot can be used for numerical or catogerical values
# values on y-axis shows in percentile format
# small circles in graph called as outliers

s.boxplot('sex','tip',data=df)
```

```python
s.boxplot('day','tip',data=df,palette='rainbow')
```

```python
s.boxplot(data=df)
```

```python
# categorize my data based on 1 feature (possible through 'hue')

s.boxplot('total_bill','day',data=df,hue='sex')
```

## Violin Plot
- Violin plot helps us to see both the distribution of data in terms of Kernel density estimation and the box plot

```python
s.violinplot('day','tip',data=df,kde=False)
```
