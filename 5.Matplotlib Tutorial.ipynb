#### Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+.
###### Some of the major Pros of Matplotlib are:
- Generally easy to get started for simple plots
- Support for custom labels and texts
- Great control of every element in a figure
- High-quality output in many formats
- Very customizable in general

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
a=np.arange(21,30)
b=np.arange(31,40)
```

```python
plt.scatter(x,y,c='r') # x and y must be the same size 
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('2D graph')
```

```python
plt.savefig('scatter_graph_example.png') #we can also save the image by using function or by right clicking the image 
```

```python
y=x*x
```

```python
plt.plot(x,y,'g*--',markersize=12)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('2D graph')
```

```python
#creating subplots

plt.subplot(2,2,1) #means 2-rows,2-columns,1-first place the below graph should be there
plt.plot(x,y,'b*--')
plt.subplot(2,2,2) #means 2-rows,2-columns,2-second place the below graph should be there
plt.plot(x,y,'r*--')
plt.subplot(2,2,3) #means 2-rows,2-columns,3-third place the below graph should be there
plt.plot(x,y,'go')
```

```python
#sine curve

x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y)
```

```python
##Bar plot

x=[2,4,6,8]
y=[3,5,7,9]

x1=[1,2,3,4]
y1=[5,6,7,8]

plt.bar(x,y)
plt.bar(x1,y1,color='g')
```

```python
## Histograms

a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]) 
plt.hist(a,bins=20) 
plt.title("histogram") 

#bins=intervals shown on x-axis
#In histogram graph we can see equal intervals on x-axis and density/number of values on y-axis
```

```python
#Box plot using Matplotlib

data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=False);
#small circles represent outliers
```

```python
#now if we see the data there are so many values so we are using box plot to represent
data
```

```python
#Pie chart

lables='covid_positive cases','covid_death cases','covid_recovery cases'
sizes=[100,25,70]
explode=(0.1,0,0) #explode(show seperately) 1st case
colors=['green','red','blue']
plt.pie(sizes,explode=explode,labels=lables,colors=colors,autopct='%1.1f%%')
#autopct show number(percentage) in pie chart
plt.axis('equal')
```
